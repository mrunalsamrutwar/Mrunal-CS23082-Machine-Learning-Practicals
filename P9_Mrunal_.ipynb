```python
inertia_values = []
for k in range(1,11):
      kmeans = KMeans(n_clusters=k, random_state=42)
      kmeans.fit(X_scaled)
      inertia_values.append(kmeans.inertia_)

      print(inertia_values)
      plt.figure()
```

```python
print(inertia_values)
plt.figure()
plt.plot(range(1,11),inertia_values,marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```
